```python
import tkinter as tk
from random import choice

#on prepare le jeu en creant la base de donnees des mots
dictionnaire = {   
    5: {
        "amour": "sentiment souvent représenté par un symbole",
        "danse": "sport qui s'accompagne souvent de musique",
        "sable": "ce qui tapisse le sol des lieux de vacances",
        "valse": "danse caractérisée par un mouvement de rotation",
        "vache": "animal de la ferme",
        "aigle":"",
        "bijou":""},

    6: {"soleil": "qui éclaire la terre",
        "tortue": "animal à carapace",
        "puzzle": "jeu d'assemblage calme",
        "bateau": "permet de voyager sur les eaux",
        "cactus": "plante que l'on trouve dans les zones désertiques",
        "animal": "",
        "balcon": "",
        "bouche": "",
        "maison": "habitat pour les humains",},

    7: {"cartable": "sac utilisé pour l'école",
        "dessert": "plat sucré mangé à la fin du repas",
        "tornade": "vent très violent",
        "village": "petit groupe de maisons",
        "trousse":"objet scolaire dans lequel on ",
        "caillou": "",
        "cadenas":""},

    8: {"aventure": "histoire avec des événements imprévus",
        "boussole": "objet qui aide à se repérer",
        "telescope": "objet pour observer les objets lointains",
        "grenouille": "animal qui saute et nage",
        "pyramide": "grand monument de forme triangulaire",
        "bracelet":"",
        "boutique":"",
        "chemise":"",
        "banquise":"grande étendue de glace",},

}

#on choisit la taille du mot
#a faire : creer un bouton (a modifier en tkinter)
longueur = int(input("saisir la longueur du mot:"))

#on choisit un mot au hasard et on recupere son indice
mot = choice(list(dictionnaire[longueur].keys()))
indice = dictionnaire[longueur][mot]


trouves = ["*"]* len(mot)  # c'est ce que voit le joueur
#Exemple :
#mot = "soleil"
#trouves = ["*", "*", "*", "*", "*", "*"]

poubelle = []   #on y stock les lettres deja utilisées
nb_erreurs = 0  #compteur nb d'erreurs
max_erreurs = 8



#INTERFACE GRAPHIQUE:
fenetre = tk.Tk()  #creation de la fenetre
fenetre.title("Pendu")  #on donne un titre a la fenetre

#LES LABELS:

#on cree un label pour ecrire "*****"
label_mot = tk.Label(fenetre, text=" ".join(trouves), font=("Arial", 20))
# on pourra utiliser grid() pour positionner apres 
label_mot.pack()

#on cree un label d'informations
label_info = tk.Label(fenetre, text="", font=("Arial", 12))
label_info.pack()

#on cree un label pour enumerer le nb d'erreurs:
label_erreurs = tk.Label(fenetre, text="erreurs = 0", font=("Arial", 12))
label_erreurs.pack()

#on cree un label pour la poubelle
label_poubelle = tk.Label(fenetre, text="lettres utilisées:", font=("Arial", 12))
label_poubelle.pack()

#on cree la zone ou l'utilisateur tape une lettre
entry = tk.Entry(fenetre)
entry.pack()
bouton = tk.Button(fenetre, text="Valider", command=valider)  #bouton qui appelle la fonction valider()
bouton.pack() 
canvas = tk.Canvas(fenetre, width="800", height="600")

canvas.pack()

#support (erreur 1)

supp1 = canvas.create_line(250,450,500,450, width=3, state="hidden")
supp2 = canvas.create_line(300,100,300,450, width=3, state="hidden")
supp3 = canvas.create_line(300,100,450,100, width=3, state="hidden")

#tête (erreur 2)
corde = canvas.create_line(450,100,450,150, width=3, state="hidden")
tete = canvas.create_oval(415,150,485,210, width=3, state="hidden")

#corps (erreur 3)
corps = canvas.create_line(450,210,450,300, width=3, state="hidden")

#jambe 1 (erreur 4)
jambe1= canvas.create_line(450,300,480,375, width=3, state="hidden")

#jambe 2 (erreur 5)
jambe2 = canvas.create_line(450,300,420,375, width=3, state="hidden")

#bras 1 (erreur 6)
bras1 = canvas.create_line(450,240,480,300, width=3, state="hidden")

#bras2 (erreur 7)
bras2 = canvas.create_line(450,240,420,300, width=3, state="hidden")
#visage erreur 8
oeil1 = canvas.create_oval(435,175,440,180, width=3, state="hidden")
oeil2 = canvas.create_oval(465,175,470,180, width=3, state="hidden")
bouche = canvas.create_line(450,200,455,200, width=3, state="hidden")

# liste pour avoir l'ordre d'affichage des dessins selon le nb d'erreurs
etapes_dessin = [
    [supp1, supp2, supp3], # Erreur 1
    [corde, tete], # Erreur 2
    [corps],       # Erreur 3
    [jambe1],      # Erreur 4
    [jambe2],      # Erreur 5
    [bras1],       # Erreur 6
    [bras2],         # Erreur 7
    [oeil1, oeil2, bouche] # Erreur 8
]


def valider():   #cette fonction s'execute a chaque clic
    global nb_erreurs
    lettre = entry.get()   #recupere ce que l'utilisateur tape
    entry.delete(0, tk.END)  #vide la case 

    if lettre in poubelle or lettre in trouves:
        label_info.config(text="lettre deja utilisee")

    elif lettre in mot:
        for i, j in enumerate(mot):
            if j == lettre:
                trouves[i] = lettre
        label_info.config(text="bonne lettre")

    else:
        if nb_erreurs < len(etapes_dessin):
            #recup des pieces a afficher pour le nb d'erreurs actuel
            pieces = etapes_dessin[nb_erreurs] 
            for p in pieces:
                canvas.itemconfig(p, state='normal') #on les affiche
        
        poubelle.append(lettre)
        label_info.config(text="mauvaise lettre")
        nb_erreurs += 1

    label_mot.config(text=" ".join(trouves)) #met a jour l'affichage du mot
    label_erreurs.config(text="Erreurs:" + str(nb_erreurs))
    label_poubelle.config(text="Lettres utilisées : " + " ".join(poubelle))
    print(trouves)    #ex: trouves = ["*", "o", "*", "*", "*", "*"]
    print(poubelle)  



fenetre.mainloop()



#ce qui reste :
#message gagné / perdu  
#dessin du pendu (canvas) fait par shama
#bouton indice
#choix de la longueur en Tkinter (au lieu de input)
#prendre en consideration si on met des lettres en majuscule ou en minuscule, ou des chiffres
```
