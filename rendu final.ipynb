```python
import tkinter as tk
from random import choice

# --- FONCTIONS DE LOGIQUE ---

def demarrer_jeu(mode):
    """Initialise le jeu selon le mode et cache le menu de sélection."""
    global etapes_dessin, jeu_actif
    
    if mode == "facile": etapes_dessin = etapes_dessinf
    elif mode == "moyen": etapes_dessin = etapes_dessinm
    else: etapes_dessin = etapes_dessind
    
    # On détruit le menu de sélection
    frame_difficulte.destroy()
    # On affiche l'interface de jeu (cachée au début)
    frame_jeu.pack(pady=20)
    jeu_actif = True

def Valider():
    global nb_erreurs, jeu_actif
    if not jeu_actif: return

    lettre = entry.get().lower() 
    entry.delete(0, tk.END)

    if not lettre or len(lettre) > 1:
        label_info.config(text="Entrez une seule lettre", fg="orange")
        return

    if lettre in poubelle or lettre in trouves:
        label_info.config(text="Lettre déjà utilisée", fg="orange")
        return

    if lettre in mot:
        for i, j in enumerate(mot):
            if j == lettre:
                trouves[i] = lettre
        label_info.config(text="Bonne lettre !", fg="green")
    else:
        if nb_erreurs < len(etapes_dessin):
            pieces = etapes_dessin[nb_erreurs] 
            for p in pieces:
                canvas.itemconfig(p, state='normal')
        
        poubelle.append(lettre)
        label_info.config(text="Mauvaise lettre...", fg="red")
        nb_erreurs += 1

    # Mise à jour affichage
    label_mot.config(text=" ".join(trouves))
    label_erreurs.config(text=f"Erreurs : {nb_erreurs} / {len(etapes_dessin)}")
    label_poubelle.config(text="Lettres utilisées : " + ", ".join(poubelle))

    # Vérification Victoire / Défaite
    if "*" not in trouves:
        fin_de_partie("GAGNÉ !", "green")
    elif nb_erreurs >= len(etapes_dessin):
        fin_de_partie(f"PERDU !\nLe mot était : {mot.upper()}", "red")

def fin_de_partie(message, couleur):
    global jeu_actif
    jeu_actif = False
    # On dessine le texte de fin directement sur le canvas au centre
    canvas.create_rectangle(150, 200, 650, 400, fill="white", outline=couleur, width=5)
    canvas.create_text(400, 300, text=message, font=("Arial", 30, "bold"), fill=couleur, justify="center")
    entry.config(state="disabled")
    bouton.config(state="disabled")

# --- DONNÉES ---
dictionnaire = {
    5: {"amour": "Sentiment", "danse": "Sport", "vache": "Animal"},
    6: {"soleil": "Astre", "tortue": "Animal", "maison": "Habitat"},
    7: {"village": "Lieu", "trousse": "Scolaire", "caillou": "Nature"},
    8: {"aventure": "Histoire", "boussole": "Objet", "banquise": "Glace"}
}

# Configuration initiale
longueur = int(input("Saisir la longueur du mot (5 à 8) : "))
mot = choice(list(dictionnaire[longueur].keys()))
trouves = ["*"] * len(mot)
poubelle = []
nb_erreurs = 0
jeu_actif = False

# --- INTERFACE GRAPHIQUE ---
fenetre = tk.Tk()
fenetre.title("Le Jeu du Pendu")
fenetre.geometry("800x800")

# 1. MENU DE DIFFICULTÉ (Apparaît en premier)
frame_difficulte = tk.Frame(fenetre)
frame_difficulte.pack(pady=50)

tk.Label(frame_difficulte, text="Choisissez votre difficulté :", font=("Arial", 16)).pack(pady=10)
tk.Button(frame_difficulte, text="Facile (8 chances)", font=("Arial", 12), width=20, command=lambda: demarrer_jeu("facile")).pack(pady=5)
tk.Button(frame_difficulte, text="Moyen (5 chances)", font=("Arial", 12), width=20, command=lambda: demarrer_jeu("moyen")).pack(pady=5)
tk.Button(frame_difficulte, text="Difficile (2 chances)", font=("Arial", 12), width=20, command=lambda: demarrer_jeu("difficile")).pack(pady=5)

# 2. ZONE DE JEU (Cachée au début)
frame_jeu = tk.Frame(fenetre)

label_mot = tk.Label(frame_jeu, text=" ".join(trouves), font=("Courier", 30, "bold"))
label_mot.pack(pady=10)

label_info = tk.Label(frame_jeu, text="Bonne chance !", font=("Arial", 12))
label_info.pack()

label_erreurs = tk.Label(frame_jeu, text="Erreurs : 0", font=("Arial", 12))
label_erreurs.pack()

label_poubelle = tk.Label(frame_jeu, text="Lettres utilisées : ", font=("Arial", 10), fg="grey")
label_poubelle.pack()

entry = tk.Entry(frame_jeu, font=("Arial", 18), width=5, justify="center")
entry.pack(pady=10)
entry.bind('<Return>', lambda e: Valider()) # Permet de valider avec la touche Entrée

bouton = tk.Button(frame_jeu, text="Valider", command=Valider, bg="lightblue")
bouton.pack()

canvas = tk.Canvas(frame_jeu, width=800, height=450, bg="white", highlightthickness=1, highlightbackground="black")
canvas.pack(pady=20)

# --- DESSINS DU PENDU ---
supp1 = canvas.create_line(250,400,550,400, width=5, state="hidden") # Socle
supp2 = canvas.create_line(300,50,300,400, width=5, state="hidden")  # Poteau
supp3 = canvas.create_line(300,50,450,50, width=5, state="hidden")   # Traverse
corde = canvas.create_line(450,50,450,100, width=3, state="hidden")
tete = canvas.create_oval(415,100,485,170, width=3, state="hidden")
corps = canvas.create_line(450,170,450,280, width=3, state="hidden")
bras1 = canvas.create_line(450,200,410,250, width=3, state="hidden")
bras2 = canvas.create_line(450,200,490,250, width=3, state="hidden")
jambe1 = canvas.create_line(450,280,410,350, width=3, state="hidden")
jambe2 = canvas.create_line(450,280,490,350, width=3, state="hidden")
# Visage
oeil1 = canvas.create_line(430,125,440,135, width=2, state="hidden")
oeil1b = canvas.create_line(440,125,430,135, width=2, state="hidden")
oeil2 = canvas.create_line(460,125,470,135, width=2, state="hidden")
oeil2b = canvas.create_line(470,125,460,135, width=2, state="hidden")
bouche = canvas.create_arc(430,140,470,160, start=0, extent=180, width=2, state="hidden")

# --- LISTES D'ÉTAPES ---
etapes_dessinf = [[supp1, supp2, supp3], [corde], [tete], [corps], [bras1], [bras2], [jambe1], [jambe2, oeil1, oeil1b, oeil2, oeil2b, bouche]]
etapes_dessinm = [[supp1, supp2, supp3, corde], [tete], [corps], [bras1, bras2], [jambe1, jambe2, oeil1, oeil1b, oeil2, oeil2b, bouche]]
etapes_dessind = [[supp1, supp2, supp3, corde, tete], [corps, bras1, bras2, jambe1, jambe2, oeil1, oeil1b, oeil2, oeil2b, bouche]]

fenetre.mainloop()
```
